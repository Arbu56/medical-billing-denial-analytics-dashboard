# Cleaning Data

```python
Data.duplicated().sum()
```

# Removed Duplicate Date

```python
Data = Data.drop_duplicates()
```

```python
Data.duplicated().sum()
```

# Deleting Extra space and in one format

```python
Data['Claim_Status'] = Data['Claim_Status'].str.strip().str.title()
```

```python
Data['Insurance_Company'] = Data['Insurance_Company'].str.strip().str.title()
```

# Changing Billed amt datatype 

```python
Data['Billed_Amount'] = pd.to_numeric(Data['Billed_Amount'], errors='coerce')
```

```python
Data.info()
```

# Filing and deleting Null values

```python
Data.isnull().sum()
```

```python
Data['Insurance_Company'] = Data['Insurance_Company'].fillna('Unknown')
```

```python
Data['Rendering_Provider'] = Data['Rendering_Provider'].fillna('Unknown provider')
```

```python
Data['blan'] = Data['blan'].fillna('Na')
```

```python
Data['Paid_Amount'] = Data['Paid_Amount'].fillna('0')
```

```python
Data['Remaining_Balance'] = Data['Remaining_Balance'].fillna('0')
```

```python
Data['Appeal_Status'] = Data['Appeal_Status'].fillna('Not Started')
```

```python
Data['Billed_Amount'] = Data['Billed_Amount'].fillna(Data['Billed_Amount'].median())
```

```python
Data.isnull().sum()
```

# Deleting unwanted column

```python
Data.columns
```

```python
Data.Appeal_Level
```

```python
Data = Data.drop(columns=['Appeal_Level','ICD10_Code'])
```

# Changing Name

```python
Data = Data.rename(columns={'blan': 'CPT'})
```

```python
Data.columns
```

```python
Data = Data.drop(columns=['Allowed_Amount'])
```

# Creating Year and Month column for comparison

```python
Data['DOS'] = pd.to_datetime(Data['DOS'])
```

```python
Data['DOS_Year'] = Data['DOS'].dt.year
```

```python
Data['DOS_Month'] = Data['DOS'].dt.month_name()
```

# Creating Denial and paid flags for easy calculation 

```python
Data['Claim_Status'].value_counts()
```

```python
Data['Denial_Flag'] = np.where(Data['Claim_Status'] == 'Denied',1,0)
```

```python
Data['Paid_Flag'] = np.where(Data['Claim_Status'] == 'Paid',1,0)
```
